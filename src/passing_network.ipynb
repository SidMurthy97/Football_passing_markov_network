```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#get raw data
passing_network = pd.read_csv(filepath_or_buffer=r"..\data\england_passing_data_vs_senegal.csv")
passing_network
```

```python
#sum up the total pass each person does
passing_network["total_passes"] = passing_network.sum(axis='columns', numeric_only=True)
passing_network
```

```python
#perform arithmetic on the absolute passing statistics and persist the passes
#from and to column

passing_network_as_percents = pd.concat([passing_network["Passes From v to >"],
    (passing_network.iloc[:,1:-1].div(passing_network["total_passes"].values,
         axis = 0)).multiply(100).round(2)], axis = 1)

passing_network_as_percents
```

```python
# Figure out how to make an edgelist with the top x connections 
cutoff_players = 3

player_data_df = pd.DataFrame()
edgelist = pd.DataFrame()

for i in range(11):
    player_data_to = passing_network_as_percents.iloc[i,1:] #data on player being passed to
    player_from = passing_network_as_percents.iloc[i,0] #name of player doing the passing

    #sort the values 
    player_data_to.sort_values(ascending=False, inplace=True)
    

    player_data_df["From"] = [player_from] * cutoff_players
    player_data_df["To"] = player_data_to.iloc[:cutoff_players].index
    player_data_df["value"] = player_data_to[:cutoff_players].values
    
    #append to higher dataframe
    edgelist = edgelist.append(player_data_df, ignore_index=True)
# player_data_df.head(cutoff_players)
edgelist
```

```python
#now to use networkx
import networkx as nx
G = nx.from_pandas_edgelist(edgelist,
                            source="From",
                            target="To",
                            edge_attr="value")
```

```python
# import pyvis
from pyvis.network import Network
# create vis network
net = Network(notebook=True, width=1000, height=600)
# load the networkx graph
net.from_nx(G)
# show
net.show("../outputs/passing_network.html")
```
