```python
import numpy as np
import pandas as pd
```

```python
#get raw data
passing_network = pd.read_csv(filepath_or_buffer=r"..\data\england_passing_data_vs_senegal.csv")
passing_network
```

```python
#sum up the total pass each person does
passing_network["total_passes"] = passing_network.sum(axis='columns', numeric_only=True)
passing_network
```

```python
#perform arithmetic on the absolute passing statistics and persist the passes
#from and to column

passing_network_as_percents = pd.concat([passing_network["Passes From v to >"],
    (passing_network.iloc[:,1:-1].div(passing_network["total_passes"].values,
         axis = 0)).multiply(100).round(2)], axis = 1)

passing_network_as_percents
```
